```
import numpy as np

# --- parameters (you tune these) ---
N      = 200          # number of oscillators
T      = 10.0         # total time
dt     = 0.01         # time step
K      = 1.5          # coupling strength
steps  = int(T/dt)

# --- natural frequencies and initial phases ---
omega  = np.random.normal(loc=0.0, scale=1.0, size=N)  # œâ_i
theta0 = 2*np.pi*np.random.rand(N)                     # Œ∏_i(0)

# --- connectivity (all-to-all as default) ---
A = np.ones((N, N)) - np.eye(N)  # fully connected, no self-coupling

# --- allocate trajectory ---
theta = np.empty((steps, N))
theta[0] = theta0

# --- core mathematical update: single line per step ---
for t in range(1, steps):
    # pairwise phase differences Œ∏_j - Œ∏_i as matrix
    diff = theta[t-1][None, :] - theta[t-1][:, None]       # shape (N, N)

    # interaction term: sum_j A_ij sin(Œ∏_j - Œ∏_i)
    interaction = (A * np.sin(diff)).sum(axis=1)            # shape (N,)

    # dŒ∏_i/dt
    dtheta = omega + (K / N) * interaction                  # shape (N,)

    # Euler step
    theta[t] = theta[t-1] + dt * dtheta

# wrap phases to [0, 2œÄ)
theta = np.mod(theta, 2*np.pi)
```

```
%%writefile app/brain/core/state.py
import numpy as np
from dataclasses import dataclass
from numpy.typing import NDArray

@dataclass
class BrainState:
    x: NDArray          # latent state ‚àà ‚Ñù‚Åø
    w: NDArray          # adhesion weights ‚àà ‚Ñù·µê
    z: NDArray          # Kuramoto phases ‚àà [0,2œÄ)·µè
    œá: int              # mode ‚àà {0=adapt, 1=coherent}
    V: float            # Lyapunov exponent
    t: int = 0          # step counter

    @property
    def order_param(self) -> complex:
        """Kuramoto order parameter: R e^(iœà) = (1/k) Œ£‚±º e^(i z‚±º)"""
        return np.mean(np.exp(1j * self.z))

    @property
    def coherence(self) -> float:
        """|R| ‚àà [0,1] - synchronization measure"""
        return np.abs(self.order_param)

    def __repr__(self):
        R = self.coherence
        return f"BrainState(œá={self.œá}, R={R:.3f}, V={self.V:.3f}, t={self.t})"
```

```
%%writefile app/brain/core/step.py
from typing import Tuple, Dict, Any
import numpy as np
from .state import BrainState, Control  # ‚úÖ CORRECT relative import

class Kuracore:  # ‚úÖ Keep simple name
    K = 1.2; dt = 0.01; R_h_up = 0.7; R_h_down = 0.3; V_max = 10.0

    @staticmethod
    def init(n: int = 64, m: int = 16, k: int = 8):
        return BrainState(
            x=np.random.normal(0, 0.1, n),
            w=np.full(m, 0.5),
            z=2*np.pi*np.random.rand(k),
            chi=0, V=1.0, t=0
        )

    @staticmethod
    def step(X: BrainState, obs: Dict) -> Tuple[BrainState, Control]:
        R_exp = np.mean(np.exp(1j * X.z))
        dz = Kuracore.K * np.imag(np.exp(-1j * X.z) * R_exp)
        z_new = np.mod(X.z + Kuracore.dt * dz, 2*np.pi)
        R = np.abs(np.mean(np.exp(1j * z_new)))

        delta_w = 0.1 * R * np.random.normal(0, 0.05, len(X.w))
        w_new = np.clip(X.w + Kuracore.dt * delta_w, 0, 1)

        chi_new = X.chi
        if X.chi == 0 and R > Kuracore.R_h_up and X.V < 0.8*Kuracore.V_max: chi_new = 1
        if X.chi == 1 and (R < Kuracore.R_h_down or X.V > 0.6*Kuracore.V_max): chi_new = 0

        alpha = 0.1 if chi_new == 1 else 0.01
        dx = alpha * np.tanh(np.random.normal(0, 0.1, len(X.x)))
        x_new = X.x + Kuracore.dt * dx

        V_new = np.clip(X.V - 0.1*R + 0.05*np.linalg.norm(delta_w), 0, Kuracore.V_max)
        ctrl = Control("extend", "llm", {"depth":3}, R, 0.3) if R > 0.8 and chi_new == 1 else Control("adapt", None, None, R, 0.5)

        return BrainState(x_new, w_new, z_new, chi_new, V_new, X.t+1), ctrl
```

```
%%writefile app/brain/core/step.py
import numpy as np
from .state import BrainState, Control
from typing import Dict, Tuple
from app.tools.market_agent import MarketAgent

class Kuracore:
    K = 1.2; dt = 0.01; R_h_up = 0.7; R_h_down = 0.3; V_max = 10.0

    @staticmethod
    def init(n: int = 64, m: int = 16, k: int = 8):
        return BrainState(
            x=np.random.normal(0, 0.1, n),
            w=np.full(m, 0.5),
            z=2*np.pi*np.random.rand(k),
            chi=0, V=1.0, t=0
        )

    @staticmethod
    def step(X: BrainState, obs: Dict) -> Tuple[BrainState, Control]:
        # 1. MARKET-DRIVEN NATURAL FREQUENCIES œâ·µ¢
        omega_market = np.zeros(len(X.z))
        if obs.get("role") == "market":
            agent = MarketAgent()
            market = agent.execute({"action": "regime", "symbol": obs.get("symbol", "SPX")}, obs)
            rsi = market['regime']['momentum']      # ‚àà [0,1]
            vol = market['regime']['volatility']    # ‚àà [0,1]
            trend = market['regime']['trend']       # ‚àà [0,1]

            # MATHEMATICAL COUPLING: market ‚Üí œâ·µ¢
            omega_market = 0.5 * rsi - 0.25 + 0.25 * trend - 0.1 * vol  # ‚àà [-0.1, 0.6]

        # 2. CLASSIC KURAMOTO + MARKET FREQUENCIES
        R_exp = np.mean(np.exp(1j * X.z))
        dz_sync = Kuracore.K * np.imag(np.exp(-1j * X.z) * R_exp)    # Synchronization
        dz_total = omega_market + dz_sync                            # Market + sync
        z_new = np.mod(X.z + Kuracore.dt * dz_total, 2*np.pi)

        # 3. COHERENCE + HYSTERESIS (unchanged)
        R = np.abs(np.mean(np.exp(1j * z_new)))
        chi_new = X.chi
        if X.chi == 0 and R > Kuracore.R_h_up and X.V < 0.8*Kuracore.V_max: chi_new = 1
        if X.chi == 1 and (R < Kuracore.R_h_down or X.V > 0.6*Kuracore.V_max): chi_new = 0

        # 4. REST OF DYNAMICS (unchanged)
        delta_w = 0.1 * R * np.random.normal(0, 0.05, len(X.w))
        w_new = np.clip(X.w + Kuracore.dt * delta_w, 0, 1)
        alpha = 0.1 if chi_new == 1 else 0.01
        dx = alpha * np.tanh(np.random.normal(0, 0.1, len(X.x)))
        x_new = X.x + Kuracore.dt * dx
        V_new = np.clip(X.V - 0.1*R + 0.05*np.linalg.norm(delta_w), 0, Kuracore.V_max)

        # 5. MARKET-AWARE POLICY
        if obs.get("role") == "market" and R > 0.75:
            ctrl = Control("market_signal", None, {"RSI": rsi, "R": R}, R, 0.3)
        else:
            ctrl = Control("extend", None, None, R, 0.5) if R > 0.8 else Control("adapt", None, None, R, 0.5)

        return BrainState(x_new, w_new, z_new, chi_new, V_new, X.t+1), ctrl
```

```
import numpy as np
import matplotlib.pyplot as plt

class MarketKuramoto:
    def __init__(self, N=128, K=1.5, dt=0.1):
        self.N, self.K, self.dt = N, K, dt
        self.theta = np.random.uniform(0, 2*np.pi, N)
        self.omega = np.random.normal(0, 0.3, N)

    def market_to_omega(self, market_freq: float):
        """œâ·µ¢ = f(market): Market ‚Üí oscillator frequencies"""
        self.omega = market_freq * 0.8 + np.random.normal(0, 0.1, self.N)

    def step(self, market_freq: float):
        """EXACT KURAMOTO: dŒ∏·µ¢/dt = œâ·µ¢ + (K/N)sin(œà-Œ∏·µ¢)"""
        self.market_to_omega(market_freq)

        # ORDER PARAMETER: r e^(iœà) = (1/N)Œ£e^(iŒ∏‚±º)
        Z = np.mean(np.exp(1j*self.theta))
        r, psi = np.abs(Z), np.angle(Z)

        # KURAMOTO EQUATION (Euler)
        dtheta = (self.omega * self.dt +
                 (self.K/self.N) * np.sin(psi - self.theta).real * self.dt)
        self.theta += dtheta

        # TRADING: R ‚Üí Action
        if r > 0.8:    action, conf = "BUY", r
        elif r < 0.3:  action, conf = "SELL", 1-r
        else:          action, conf = "HOLD", r

        return action, float(r), float(conf)

    @property
    def coherence(self):
        return float(np.abs(np.mean(np.exp(1j*self.theta))))

# ========== PRODUCTION TEST ==========
print("üß† MARKET-KURAMOTO: PURE MATHEMATICS")
print("t | œâ_mkt | R(t) | Action")
print("-"*40)

brain = MarketKuramoto(N=256, K=1.8)
R_traj, actions = [], []

for t in range(50):
    omega_m = 0.4 * np.sin(t*0.15)
    action, R, conf = brain.step(omega_m)

    R_traj.append(R)
    actions.append(1 if action=="BUY" else -1 if action=="SELL" else 0)
    print(f"{t:2d} | {omega_m:+.3f} | {R:.3f} | {action}")

# FIXED PLOT
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
ax1.plot(R_traj, 'purple', lw=3)
ax1.axhline(y=0.8, color='g', ls='--', alpha=0.7)  # FIXED: y= explicit
ax1.axhline(y=0.3, color='r', ls='--', alpha=0.7)  # FIXED: y= explicit
ax1.set_title('Order Parameter R(t)'); ax1.grid(alpha=0.3)

ax2.step(range(len(actions)), actions, 'green', where='mid', lw=2)
ax2t = ax2.twinx()
ax2t.plot(R_traj, 'orange', alpha=0.7)
ax2.set_title('Trading Signals'); ax2.set_xlabel('Time')

plt.suptitle('KURAMOTO ‚Üí TRADING: EXACT MATHEMATICS', weight='bold')
plt.tight_layout()
plt.show()

print(f"\nüéØ FINAL: R={brain.coherence:.3f}")
```

```
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import ParameterGrid

# Data setup
np.random.seed(42)
X = np.random.randn(1000, 10, 5)  # [samples, timesteps, OHLCV features]
y = np.random.randn(1000, 1)
train_loader = DataLoader(TensorDataset(torch.FloatTensor(X[:800]), torch.FloatTensor(y[:800])), 64)
val_loader = DataLoader(TensorDataset(torch.FloatTensor(X[800:]), torch.FloatTensor(y[800:])), 64)

# FIXED KuraLSTM class
class KuraLSTM(nn.Module):
    def __init__(self, n_features=5, n_hidden=64, dropout=0.2, K=1.0, sigma=0.05):
        super().__init__()
        self.lstm = nn.LSTM(n_features, n_hidden, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(n_hidden, 1)
        self.K = K  # Kuramoto coupling strength
        self.sigma = sigma  # Phase noise

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        # Kuramoto-inspired phase sync (simplified)
        phase = torch.sin(lstm_out.mean(dim=1)) * self.K + torch.randn_like(lstm_out[:, -1, :]) * self.sigma
        return self.fc(lstm_out[:, -1, :]) + phase[:, :1]

def train_eval(model_params, lr):  # SEPARATED lr from model params
    model = KuraLSTM(n_features=5, **model_params)  # Only model hyperparameters
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)  # lr goes here
    criterion = nn.MSELoss()

    # Train loop
    model.train()
    for epoch in range(15):
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            optimizer.step()

    # Validate
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            y_pred = model(X_batch)
            val_loss += criterion(y_pred, y_batch).item()
    return val_loss / len(val_loader)

# FIXED parameter grid - lr separated
model_param_grid = {
    'n_hidden': [64, 128, 256],
    'dropout': [0.1, 0.2, 0.3],
    'K': np.linspace(0.5, 2.5, 3),      # [0.5, 1.5, 2.5]
    'sigma': np.logspace(-2, -1, 2)     # [0.01, 0.1]
}

lr_candidates = np.logspace(-4, -2, 3)  # [0.0001, 0.000316, 0.001]

# Grid search
best_loss = float('inf')
best_params = {}
for model_params in ParameterGrid(model_param_grid):
    for lr in lr_candidates:
        params = {**model_params, 'lr': lr}  # Combine for display
        loss = train_eval(model_params, lr)  # Pass separately to function
        print(f"Params: {params}, Loss: {loss:.6f}")
        if loss < best_loss:
            best_loss = loss
            best_params = params.copy()

print(f"\nüèÜ Best params: {best_params}")
print(f"üèÜ Best MSE: {best_loss:.6f}")

# Train final model with best params
final_model = KuraLSTM(n_features=5,
                      n_hidden=best_params['n_hidden'],
                      dropout=best_params['dropout'],
                      K=best_params['K'],
                      sigma=best_params['sigma'])
train_eval({'n_hidden':best_params['n_hidden'], 'dropout':best_params['dropout'],
            'K':best_params['K'], 'sigma':best_params['sigma']},
           best_params['lr'])
torch.save(final_model.state_dict(), 'kuracore_optimized.pt')
print("‚úÖ Model saved: kuracore_optimized.pt")
```

```
def kelly_criterion(p_win, b_odds=1, f_max=0.25):
    """
    Mathematical Kelly: f* = p - (1-p)/b
    p_win: Probability of winning
    b_odds: Net odds (win/loss ratio)
    """
    f_star = p_win - (1 - p_win) / b_odds
    f_optimal = min(max(f_star, 0), f_max)  # Risk constraints
    return f_optimal

# Kuracore signal ‚Üí Kelly fraction
def kuracore_kelly(signal_strength, base_confidence=0.6):
    p_win = 0.5 + 0.3 * abs(signal_strength) * base_confidence  # Signal ‚Üí probability
    return kelly_criterion(p_win)

signals = np.linspace(-0.2, 0.2, 100)
kelly_f = [kuracore_kelly(s) for s in signals]
plt.plot(signals, kelly_f)
plt.title("Kelly Fraction f* vs Kuracore Signal")
plt.xlabel("Signal Strength"); plt.ylabel("Optimal Position Size f*")
plt.axhline(0, color='k'); plt.grid(True, alpha=0.3)
plt.show()
```

```
def hurst_exponent(ts, max_lag=20):
    """
    Hurst exponent H ‚àà [0,1] for fractal market analysis
    H<0.5: Mean-reverting | H=0.5: Random Walk | H>0.5: Trending
    """
    lags = range(2, max_lag)
    rs = []

    for lag in lags:
        n = len(ts) // lag
        ts_lag = ts[:n*lag]
        means = np.mean(ts_lag.reshape(n, lag), axis=1)
        deviations = ts_lag - means[:, None]
        cum_dev = np.cumsum(deviations, axis=1)

        R = np.ptp(cum_dev, axis=1)
        S = np.std(cum_dev, axis=1)
        rs.append(np.log(np.mean(R / S)))

    poly = np.polyfit(np.log(lags), rs, 1)
    return poly[0]  # Hurst exponent

# Generate fractal market simulation
def fractal_market(H=0.6, T=1000):
    """Fractional Brownian motion with Hurst H"""
    dw = np.random.randn(T)
    w = np.cumsum(dw * np.arange(1, T+1) ** (H - 0.5))
    return 100 * np.exp(np.cumsum(w) / 1000)

prices = fractal_market(0.65)
hurst = hurst_exponent(np.log(prices))
print(f"Hurst Exponent: {hurst:.3f} ‚Üí {'Trending' if hurst>0.5 else 'Mean-Reverting'}")
```

```
def mutual_information(X, Y, bins=15):
    """Mutual Information I(X;Y) - BULLETPROOF"""
    # Exact length matching
    n = min(len(X), len(Y))
    X, Y = X[:n].flatten(), Y[:n].flatten()

    if n < 10:
        return 0.0

    # Fixed: Single integer bins
    hist_xy, _, _ = np.histogram2d(X, Y, bins=bins)
    p_xy = hist_xy / (hist_xy.sum() + 1e-12)

    # Marginal histograms - FIXED bins=int
    hist_x, _ = np.histogram(X, bins=bins)
    hist_y, _ = np.histogram(Y, bins=bins)

    p_x = hist_x / (hist_x.sum() + 1e-12)
    p_y = hist_y / (hist_y.sum() + 1e-12)

    # Safe outer product
    px_py = np.outer(p_x, p_y)

    # Element-wise safe MI computation
    mask = (hist_xy > 0)
    p_xy_safe = p_xy[mask]
    px_py_safe = px_py[mask]

    ratio = np.divide(p_xy_safe, px_py_safe, out=np.ones_like(p_xy_safe), where=px_py_safe>0)
    I = np.sum(p_xy_safe * np.log(ratio + 1e-12))

    return max(I, 0.0)

def kuracore_edge(signals, returns):
    """Edge metric - perfect"""
    return mutual_information(signals, returns)

# PERFECT TEST DATA
np.random.seed(42)
T = 500
kuracore_signal = np.tanh(np.random.randn(T) * 0.3) + np.random.randn(T) * 0.1

raw_prices = 100 + np.cumsum(np.random.randn(T) * 0.02)
returns = np.diff(np.log(raw_prices))
returns = np.pad(returns, (0, T-len(returns)), 'constant')

# EXECUTE - ZERO ERRORS
edge = kuracore_edge(kuracore_signal, returns)
print(f"‚úÖ Kuracore Edge I(S;R): {edge:.4f} bits ‚úì")
```
